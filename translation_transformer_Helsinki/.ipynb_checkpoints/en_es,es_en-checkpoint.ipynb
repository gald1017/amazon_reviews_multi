```python
import pandas as pd
import numpy as np
import time
import torch
```

**Data sets**

### translated English to Spanish Helsinki

```python
eng_ds_train = pd.read_csv('../data/train_en')
es_ds_train = pd.read_csv('../data/train_es')
de_ds_train = pd.read_csv('../data/train_de')

eng_ds_val = pd.read_csv('../data/val_en')
es_ds_val = pd.read_csv('../data/val_es')
de_ds_val = pd.read_csv('../data/val_de')

eng_ds_test = pd.read_csv('../data/test_en')
es_ds_test = pd.read_csv('../data/test_es')
de_ds_test = pd.read_csv('../data/test_de')
```

**Translate spanish to english**

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = 'Helsinki-NLP/opus-mt-es-en'
# Download the model and the tokenizer
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
                
```

```python
def translate(model, tokenizer, text):
      with torch.no_grad():
        text = text[0:509]        
        tokens = tokenizer(text, return_tensors="pt", padding=True).to('cuda')
        tokens["input_ids"] = tokens["input_ids"][:,:512]
        tokens["attention_mask"] = tokens["attention_mask"][:, :512]
        translated = model.generate(**tokens).to('cuda')
        return [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```

```python
import pickle
import itertools

texts = [es_ds_train.review_body, es_ds_val.review_body, es_ds_test.review_body]
files_name = ['amz_train_es_to_en_Translation_Helsinki.pkl', 'amz_val_es_to_en_Translation_Helsinki.pkl', 'amz_test_es_to_en_Translation_Helsinki.pkl' ]
model.to('cuda')

for i,t in enumerate(texts):
    translation = []
    start = time.time()
    for com in t:
        translation.append(translate(model,tokenizer,com))
        print(time.time())
    with open(files_name[i], 'wb') as f:
        torch.save(list(itertools.chain(*translation)), f)    
    end = time.time()
    print(end - start)
```
